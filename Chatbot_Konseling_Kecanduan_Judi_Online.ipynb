```python
from langchain_community.document_loaders import PyPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import Chroma
from langchain.chains import RetrievalQA
from langchain_google_genai import ChatGoogleGenerativeAI
import gradio as gr
```

```python
from google.colab import files
uploaded = files.upload()
pdf_path = next(iter(uploaded))
```

```python
def buat_chain_dari_pdf(pdf_path):
    loader = PyPDFLoader(pdf_path)
    dokumen = loader.load()
    splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
    chunks = splitter.split_documents(dokumen)
    embeddings = HuggingFaceEmbeddings(model_name='sentence-transformers/all-MiniLM-L6-v2')
    vectorstore = Chroma.from_documents(chunks, embedding=embeddings)
    llm = ChatGoogleGenerativeAI(model='gemini-1.5-flash', temperature=0.3)
    chain = RetrievalQA.from_chain_type(llm=llm, retriever=vectorstore.as_retriever())
    return chain
```

```python
qa_chain = buat_chain_dari_pdf("dataset_judi.pdf")
```

```python
# Fungsi chatbot konseling dengan solusi berbasis PDF
def chat(query, history=[]):
    try:
        prompt = f"""
        Kamu adalah seorang konselor profesional yang ahli dalam membantu orang yang mengalami kecanduan judi online.
        Tugasmu adalah menjawab pertanyaan atau keluhan pengguna dengan empati, tanpa menghakimi, dan memberikan solusi nyata.

        Berikut adalah keluhan atau pertanyaan pengguna:
        "{query}"

        Jawablah dalam dua bagian:
        1. Identifikasi Masalah Singkat
        2. Solusi atau Saran (minimal 5 kalimat, maksimal 10 kalimat) yang relevan, bernuansa menenangkan, dan realistis.

        Format:
        Masalah: ...
        Solusi: ...
        """
        result = qa_chain.run(prompt)
        history.append((query, result.strip()))
    except Exception as e:
        history.append((query, f"Terjadi kesalahan: {str(e)}"))
    return history, history


# ğŸ’¬ Antarmuka Gradio gaya WhatsApp untuk Chatbot Konseling Judi
with gr.Blocks(css="""
    .gradio-container { font-family: 'Segoe UI'; }
    .chatbot .message.user {
        background-color: #a5d6a7;  /* hijau lembut */
        text-align: right;
        color: #000000 !important;
    }
    .chatbot .message.bot {
        background-color: #1565c0;  /* biru tua */
        border: 1px solid #0d47a1;
        text-align: left;
        color: #ffffff !important;
    }
    .chatbot {
        font-size: 16px;
    }
""") as demo:

    gr.Markdown("### ğŸ’¬ Konseling AI - Kecanduan Judi Online ğŸ¯")
    chatbot = gr.Chatbot(label="Sesi Konseling", elem_classes="chatbot")
    with gr.Row():
        msg = gr.Textbox(placeholder="Tulis pertanyaan atau curhatanmu di sini...", label="Pertanyaan", scale=6)
        send_btn = gr.Button("Kirim", scale=1)
    clear_btn = gr.Button("Bersihkan Obrolan")
    state = gr.State([])

    def chat_konseling(query, history=[]):
        try:
            prompt = f'''
            Kamu adalah konselor profesional yang membantu pengguna mengatasi kecanduan judi online.
            Tanggapi dengan empati, gunakan bahasa yang menenangkan dan tidak menghakimi, serta berikan solusi konkrit.

            Curhatan/Pertanyaan: {query}
            '''
            result = qa_chain.run(prompt)
            history.append((query, result.strip()))
        except Exception as e:
            history.append((query, f"Terjadi kesalahan: {str(e)}"))
        return history, history

    send_btn.click(chat_konseling, [msg, state], [chatbot, state])
    msg.submit(chat_konseling, [msg, state], [chatbot, state])
    clear_btn.click(lambda: ([], []), None, [chatbot, state])

demo.launch()
```
